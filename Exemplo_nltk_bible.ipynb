# Importar bibliotecas

```python
import nltk    
from nltk import tokenize
```

```python
from nltk.probability import FreqDist
```

```python
import re
```

```python
from pathlib import Path
```

## Carregar arquivo

```python
texto = Path("livros/livro_the_king_james_bible.txt").read_text().replace('\n', ' ')
```

```python
type(texto)
```

## Pré processamento

```python
def pre_processamento(texto):
  
    # seleciona apenas letras e coloca todas em minúsculo 
    letras_min =  re.findall(r'\b[A-zÀ-úü]+\b', texto.lower())

    # remove stopwords
    stopwords = nltk.corpus.stopwords.words()
    stop = set(stopwords)
    sem_stopwords = [w for w in letras_min if w not in stop]

    # juntando os tokens novamente em formato de texto
    texto_limpo = " ".join(sem_stopwords)

    return texto_limpo
```

```python
texto_processado = pre_processamento(texto)
```

```python
texto_tokenizado = tokenize.word_tokenize(texto_processado) 
```

## Criar texto plano

```python
texto_plano = nltk.Text(texto_tokenizado)
```

## Análise frequência

```python
freq_palavras = FreqDist(texto_tokenizado)
```

```python
print("20 palavras mais frequentes:")
print(freq_palavras.most_common(20))
```

```python
# plot
import matplotlib.pyplot as plt
plt.figure(figsize = (13, 8))
freq_palavras.plot(30, title = "Frequência de Palavras")
```

## Análise concordância

```python
texto_plano.concordance('god')
```

## Análise similaridade

```python
texto_plano.similar('god')
```
